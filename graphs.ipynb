```python
import csv
import subprocess as sb
import matplotlib.pyplot as plt
import settings
```

```python
def get_balls():
    res_arr = []
    with open(settings.RANKED_USER_FILE_PATH) as File:
        reader = csv.DictReader(File, delimiter=',')
        for row in reader:
            res_arr.append(row)
    return res_arr
```

```python
def get_users():
    res_arr = []
    with open(settings.USER_DATA_FILE_PATH) as File:
        reader = csv.DictReader(File, delimiter='|')
        for row in reader:
            res_arr.append(row)
    return res_arr
```

```python
def plot_sex_diagram(users):
    # 1 ‚Äî –∂–µ–Ω—Å–∫–∏–π;
    # 2 ‚Äî –º—É–∂—Å–∫–æ–π;
    # 0 ‚Äî –ø–æ–ª –Ω–µ —É–∫–∞–∑–∞–Ω. 
    man_count = 0
    woomen_count = 0
    not_count = 0
    for row in users:
        if(row['sex'] == '0'):
            not_count = not_count+ 1
        if(row['sex'] == '1'):
            woomen_count = woomen_count+ 1
        if(row['sex'] == '2'):
            man_count = man_count+ 1
    labels = ['male', 'female', 'not specified']
    values = [man_count, woomen_count, not_count]
    
    fig1, ax1 = plt.subplots()
    fig1.set_figwidth(6)    #  —à–∏—Ä–∏–Ω–∞ Figure
    fig1.set_figheight(6)    #  –≤—ã—Å–æ—Ç–∞ Figure
    # ax1.pie(values, labels=labels)
    # ax1.axis('equal')
    colors = ( "lightblue", "thistle", "black") 
    wedges, texts, autotexts = ax1.pie(values, labels=labels, colors = colors, autopct='%1.2f%%')
    ax1.axis('equal')
    # ax1.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
    print(f'male = {man_count}\nfemale = {woomen_count}\nnot specified = {not_count}')
    plt.show()
    
```

```python
def plot_age_diagram(users):
    count_0_17 = 0
    count_18_24 = 0
    count_25_35 = 0
    coint_36_45 = 0
    count_46_60 = 0
    count_61_100 = 0
    count_not = 0
    
    for row in users:
        age = int(row['age_users'])
        if(age < 18) and (age > 0 ):
            count_0_17 = count_0_17+ 1
        if(age >= 18) and (age<25):
            count_18_24 = count_18_24+ 1
        if(age >= 25) and (age<36):
            count_25_35 = count_25_35+ 1
        if(age >= 36) and (age<45):
            coint_36_45 = coint_36_45+ 1
        if(age >= 45) and (age<61):
            count_46_60 = count_46_60+ 1
        if(age >= 61):
            count_61_100 = count_61_100+ 1
        if age <= 0:
            count_not = count_not+1
            
    labels = ['0-17', '18-24', '25-35', '36-45', '45-60', 'more 60',  'not specified']
    values = [count_0_17, count_18_24, count_25_35, coint_36_45, count_46_60, count_61_100, count_not]
    
    
    fig1, ax1 = plt.subplots()
    fig1.set_figwidth(10)    #  —à–∏—Ä–∏–Ω–∞ Figure
    fig1.set_figheight(10)    #  –≤—ã—Å–æ—Ç–∞ Figure
    ax1.set_facecolor('seashell')
    # ax1.pie(values, labels=labels)
    # ax1.axis('equal')
    colors = ( "lightblue", "thistle", "lightgreen", 'yellow', 'wheat', 'lightgrey', 'darksalmon') 
    wedges, texts, autotexts = ax1.pie(values, labels=labels, colors = colors, autopct='%1.2f%%')
    ax1.axis('equal')
    # ax1.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
    print(f'0-17 = {count_0_17}\n18-24 = {count_18_24}\n25-35 = {count_25_35}\n36-45 = {coint_36_45}\n45-60 = {count_46_60}\nmore 60 = {count_61_100}\nnot specified = {count_not}')

    plt.show()
```

```python
def plot_city_diagram(users):

    count_spb = 0
    count_len = 0
    count_myr = 0
    count_prioz = 0
    count_gatch = 0
    count_k = 0
    count_sestr = 0
    count_moscov = 0
    count_novosib = 0
    count_kalin = 0
    count_krasnodar=0
    count_rostov = 0
    count_not = 0
    
    for row in users:
        if(row['city'] == '–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥'):
            count_spb = count_spb+ 1
        elif(row['city'] == '–õ–µ–Ω. –û–±–ª–∞—Å—Ç—å') or (row['city'] == '–õ–µ–Ω–∏–Ω–≥—Ä–∞–¥—Å–∫–∞—è –æ–±–ª–∞—Å—Ç—å'):
            count_len = count_len+ 1
        elif(row['city'] == '–ú—É—Ä–∏–Ω–æ'):
            count_myr = count_myr+ 1
        elif(row['city'] == '–ü—Ä–∏–æ–∑–µ—Ä—Å–∫'):
            count_prioz = count_prioz + 1
        elif(row['city'] == '–ì–∞—Ç—á–∏–Ω–∞'):
            count_gatch = count_gatch+ 1
        elif(row['city'] == '–ö–∏—Ä–∏—à–∏'):
            count_k = count_k+ 1
        elif(row['city'] == '–°–µ—Å—Ç—Ä–æ—Ä–µ—Ü–∫'):
            count_sestr = count_sestr+ 1
        elif(row['city'] == '–ú–æ—Å–∫–≤–∞'):
            count_moscov = count_moscov+ 1
        elif(row['city'] == '–ù–æ–≤–æ—Å–∏–±–∏—Ä—Å–∫'):
            count_novosib = count_novosib+ 1
        elif(row['city'] == '–ö–∞–ª–∏–Ω–∏–Ω–≥—Ä–∞–¥'):
            count_kalin = count_kalin+ 1
        elif(row['city'] == '–ö—Ä–∞—Å–Ω–æ–¥–∞—Ä'):
            count_krasnodar = count_krasnodar+ 1
        elif(row['city'] == '–†–æ—Å—Ç–æ–≤-–Ω–∞-–î–æ–Ω—É'):
            count_rostov = count_rostov+ 1
        else:
            count_not = count_not + 1

    labels = ['–ú–æ—Å–∫–≤–∞', '–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥',  '–ù–æ–≤–æ—Å–∏–±–∏—Ä—Å–∫', '–ö—Ä–∞—Å–Ω–æ–¥–∞—Ä', '–†–æ—Å—Ç–æ–≤-–Ω–∞-–î–æ–Ω—É', '–ö–∞–ª–∏–Ω–∏–Ω–≥—Ä–∞–¥',  '–ì–∞—Ç—á–∏–Ω–∞',  '–ö–∏—Ä–∏—à–∏', '–ü—Ä–∏–æ–∑–µ—Ä—Å–∫',
                      '–ú—É—Ä–∏–Ω–æ', 
                      '–°–µ—Å—Ç—Ä–æ—Ä–µ—Ü–∫', 
                     # '–õ–µ–Ω. –æ–±–ª–∞—Å—Ç—å', 
                     '–î—Ä—É–≥–æ–µ']
    
    x = labels
    y = [   count_moscov , count_spb,  count_novosib , count_krasnodar,  count_rostov ,  count_kalin , count_gatch ,  count_k ,  count_prioz ,
    count_myr ,
    count_sestr ,
       #  count_len ,
    count_not]

    fig, ax = plt.subplots()

    ax.bar(x, y)

    ax.set_facecolor('seashell')
    fig.set_facecolor('floralwhite')
    fig.set_figwidth(22)    #  —à–∏—Ä–∏–Ω–∞ Figure
    fig.set_figheight(6)    #  –≤—ã—Å–æ—Ç–∞ Figure


    # wedges, texts, autotexts = ax1.pie(values, labels=labels, colors = colors, autopct='%1.2f%%')


    print(f'–ú–æ—Å–∫–≤–∞ = {count_moscov}\n–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥ = {count_spb}\n–ù–æ–≤–æ—Å–∏–±–∏—Ä—Å–∫ = {count_novosib}\n–ö—Ä–∞—Å–Ω–æ–¥–∞—Ä = {count_krasnodar}\n–†–æ—Å—Ç–æ–≤-–Ω–∞-–î–æ–Ω—É = {count_rostov}\n–ö–∞–ª–∏–Ω–∏–Ω–≥—Ä–∞–¥ = {count_kalin}\n–ì–∞—Ç—á–∏–Ω–∞ = {count_gatch}\n–ö–∏—Ä–∏—à–∏ = {count_k}\n–ü—Ä–∏–æ–∑–µ—Ä—Å–∫ = {count_prioz}\n–ú—É—Ä–∏–Ω–æ = {count_myr}\n–°–µ—Å—Ç—Ä–æ—Ä–µ—Ü–∫ = {count_sestr}\n–î—Ä—É–≥–æ–µ = {count_not}')
    plt.show()


    fig1, ax1 = plt.subplots()

    x2 = ['–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥ –∏ –æ–±–ª–∞—Å—Ç—å', '–ú–æ—Å–∫–≤–∞', '–î—Ä—É–≥–æ–µ']
    y2 = [  count_spb+ count_gatch +  count_k +  count_prioz + count_myr + count_sestr , 
    count_moscov , 
    count_not+count_novosib+count_kalin+count_krasnodar+count_rostov]

    ax1.bar(x2, y2)

    ax1.set_facecolor('seashell')
    fig1.set_facecolor('floralwhite')
    fig1.set_figwidth(7)    #  —à–∏—Ä–∏–Ω–∞ Figure
    fig1.set_figheight(6)    #  –≤—ã—Å–æ—Ç–∞ Figure
    plt.show()

```

```python
def plot_bals_diagram(bols):
    total_arr = []
    zero_arr = []
    count_arr = []
    for i in range (0, 51):
        count_arr.append(0)

    for row in bols:
        mark = int(row['total_mark'])
        total_arr.append(mark)
        zero_arr.append(0)
        count_arr[mark] = count_arr[mark] + 1
    
    res_arr_mark = []
    res_arr_count=[]
    for i in range (0, 51):
        if(count_arr[i]!=0):
            res_arr_mark.append(i)
            res_arr_count.append(count_arr[i])

    res_arr_mark2 = []
    res_arr_count2=[]
    for i in range (15, 25):
        if(count_arr[i] !=0):
            res_arr_mark2.append(i)
            res_arr_count2.append(count_arr[i])

    fig, ax = plt.subplots(figsize=(10, 6))
    ax.scatter(x = total_arr , y=zero_arr)
    plt.show()
    
    fig2, ax2 = plt.subplots(figsize=(10, 6))
    ax2.plot(res_arr_mark, res_arr_count)
    plt.xlabel("–∫–æ–ª–ª–∏—á–µ—Å—Ç–≤–æ –±–∞–ª–ª–æ–≤")
    plt.ylabel("–∫–æ–ª–ª–∏—á–µ—Å—Ç–≤–æ —á–µ–ª–æ–≤–µ–∫")
    plt.show()
    
    fig3, ax3 = plt.subplots(figsize=(10, 6))
    ax3.plot(res_arr_mark2, res_arr_count2)
    plt.xlabel("–∫–æ–ª–ª–∏—á–µ—Å—Ç–≤–æ –±–∞–ª–ª–æ–≤")
    plt.ylabel("–∫–æ–ª–ª–∏—á–µ—Å—Ç–≤–æ —á–µ–ª–æ–≤–µ–∫")
    plt.show()
    
    
plot_bals_diagram(bols)
```
